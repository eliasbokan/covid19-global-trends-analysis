```python
# Data exploration
# Task 6B. Generating pseudo-random data and plot its distribution
# Import necessary libraries
import math
import random
from collections import Counter
import matplotlib.pyplot as plt

# Function to assign a point to the next lowest multiple of bucket_size
def bucketize(point, bucket_size):
    return bucket_size * math.floor(point / bucket_size)

# Function to create a histogram dictionary
def make_histogram(points, bucket_size):
    return Counter(bucketize(point, bucket_size) for point in points)

# Function to plot a histogram without lines
def plot_histogram(points, bucket_size, title=""):
    histogram = make_histogram(points, bucket_size)
    # Set figure size
    plt.figure(figsize=(8, 5))
    # Plot histogram with no edge lines and change the color to lightcoral
    plt.bar(histogram.keys(), histogram.values(), width=bucket_size, color="lightcoral")
    # Labels and title
    plt.xlabel("Value Range")
    plt.ylabel("Frequency")
    plt.title(title)
    # Show plot
    plt.show()

# Function to compute the normal cumulative distribution function (CDF)
def normal_cdf(x, mu=0, sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2

# Function to approximate the inverse of the normal CDF using binary search
def inverse_normal_cdf(p, mu=0, sigma=1, tolerance=0.00001):
    """Find approximate inverse using binary search."""
    if mu != 0 or sigma != 1:
        return mu + sigma * inverse_normal_cdf(p, mu=0, sigma=1, tolerance=tolerance)
    low_z, low_p = -10.0, 0
    hi_z, hi_p = 10.0, 1
    while hi_z - low_z > tolerance:
        mid_z = (low_z + hi_z) / 2
        mid_p = normal_cdf(mid_z)
        if mid_p < p:
            low_z, low_p = mid_z, mid_p
        elif mid_p > p:
            hi_z, hi_p = mid_z, mid_p
        else:
            break
    return mid_z

# Set seed for reproducibility
random.seed(0)

# Generate uniform distribution data in range [-100, 100]
uniform = [200 * random.random() - 100 for _ in range(10000)]
plot_histogram(uniform, 10, "Uniform Distribution Histogram")

# Generate normal distribution data with mean 0 and standard deviation 57
normal = [57 * inverse_normal_cdf(random.random()) for _ in range(10000)]
plot_histogram(normal, 10, "Normal Distribution Histogram")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Convert 'Date' column to datetime format
df_grouped["Date"] = pd.to_datetime(df_grouped["Date"])

# Step 3: Global Trends Visualization
# Global COVID-19 Trends (Grouped by Date)
global_trends = df_grouped.groupby("Date")[["Confirmed", "Deaths", "Recovered"]].sum()

# Create the plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=global_trends, linewidth=2.5)

# Add title and labels
plt.title("Global COVID-19 Trends Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Count (millions)")

# Add legend
plt.legend(["Confirmed Cases", "Deaths", "Recovered"])

# Add grid for readability
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Step 4: Top Countries by Total Cases and Deaths
# Top 10 countries by total cases and deaths
top_cases = df_covid.sort_values(by="TotalCases", ascending=False).head(10)
top_deaths = df_covid.sort_values(by="TotalDeaths", ascending=False).head(10)

# Bar plot for Top 10 countries by Total COVID-19 Cases
plt.figure(figsize=(12, 5))
sns.barplot(x="TotalCases", y="Country/Region", hue="Country/Region", data=top_cases, palette="Blues_r", legend=False)
plt.title("Top 10 Countries by Total COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()

# Bar plot for Top 10 countries by Total COVID-19 Deaths
plt.figure(figsize=(12, 5))
sns.barplot(x="TotalDeaths", y="Country/Region", hue="Country/Region", data=top_deaths, palette="Reds_r", legend=False)
plt.title("Top 10 Countries by Total COVID-19 Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("Country")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Step 5: Recovered Cases Trend
# Global Recoveries Trend
recovered_trends = df_grouped.groupby("Date")["Recovered"].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=recovered_trends.index, y=recovered_trends, color="orange")  # Changed color to orange
plt.title("Global COVID-19 Recoveries Over Time")
plt.xlabel("Date")
plt.ylabel("Recovered Cases")
plt.grid(True)
plt.show()
```

```python
# Step 6: Distribution of Total Cases and Deaths (Histograms)
# Histogram for Total Cases
plt.figure(figsize=(12, 6))
sns.histplot(df_covid["TotalCases"], bins=30, kde=True, color="purple")  # Changed color to purple
plt.title("Distribution of Total COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Histogram for Total Deaths
plt.figure(figsize=(12, 6))
sns.histplot(df_covid["TotalDeaths"], bins=30, kde=True, color="green")  # Changed color to green
plt.title("Distribution of Total COVID-19 Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
